```python
import pandas as pd
import unidecode
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar las dos bases de datos
df_2006_2018 = pd.read_excel("/content/drive/MyDrive/Ciencia de datos/corte 2/COLOMBIA/datos/datos_2006_2018.xlsx")
df_2019_2024 = pd.read_excel("/content/drive/MyDrive/Ciencia de datos/corte 2/COLOMBIA/datos/datos_2019_2024.xlsx")

print("Bases de datos cargadas.")

# 2. Renombrar columnas en df_2019_2024 para que coincidan con df_2006_2018
column_mapping = {
    'Código Dane departamento': 'CÓD. \nDEP.',
    'Departamento': 'DEPARTAMENTO',
    'Código Dane municipio': 'CÓD. MUN.',
    'Municipio': 'MUNICIPIO',
    'Grupo cultivo': 'GRUPO \nDE CULTIVO',
    'Subgrupo': 'SUBGRUPO \nDE CULTIVO',
    'Cultivo': 'CULTIVO',
    'Desagregación cultivo': 'DESAGREGACIÓN REGIONAL Y/O SISTEMA PRODUCTIVO',
    'Año': 'AÑO',
    'Periodo': 'PERIODO',
    'Área sembrada (ha)': 'Área Sembrada\n(ha)',
    'Área cosechada (ha)': 'Área Cosechada\n(ha)',
    'Producción (t)': 'Producción\n(t)',
    'Rendimiento (t/ha)': 'Rendimiento\n(t/ha)',
    'Estado físico del cultivo': 'ESTADO FISICO PRODUCCION',
    'Nombre científico del cultivo': 'NOMBRE \nCIENTIFICO',
    'Ciclo del cultivo': 'CICLO DE CULTivo' # Corrected mapping
}

df_2019_2024_renamed = df_2019_2024.rename(columns=column_mapping)

print("Columnas renombradas en df_2019_2024.")

# 3. Concatenar las dos bases de datos
df_combined = pd.concat([df_2006_2018, df_2019_2024_renamed], ignore_index=True)

print("DataFrames concatenados.")

# 4. Limpiar las columnas 'DEPARTAMENTO' y 'CULTIVO'
def clean_text(text):
    if isinstance(text, str):
        text = text.title()
        text = unidecode.unidecode(text)
    return text

df_combined['DEPARTAMENTO'] = df_combined['DEPARTAMENTO'].apply(clean_text)
df_combined['CULTIVO'] = df_combined['CULTIVO'].apply(clean_text)

print("Columnas 'DEPARTAMENTO' y 'CULTIVO' limpiadas.")


# 5. Desarrollar análisis exploratorio de los datos

# Cantidad de filas y columnas
print("\n--- Análisis Exploratorio de Datos ---")
print("Dimensiones del DataFrame combinado (filas, columnas):", df_combined.shape)

# Cantidad de datos faltantes por columna
print("\nValores faltantes por columna:")
print(df_combined.isnull().sum())

# Cantidad de cultivos únicos
num_unique_cultivos = df_combined['CULTIVO'].nunique()
print(f"\nNúmero de cultivos únicos: {num_unique_cultivos}")

# Top 10 cultivos con mayor área sembrada
top_10_cultivos_sembrada = df_combined.groupby('CULTIVO')['Área Sembrada\n(ha)'].sum().nlargest(10)

print("\nTop 10 cultivos por Área Sembrada (ha):")
print(top_10_cultivos_sembrada)

# Gráfico de los 10 cultivos con mayor área sembrada
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cultivos_sembrada.index, y=top_10_cultivos_sembrada.values, palette='viridis')
plt.title('Top 10 Cultivos por Área Sembrada (ha)')
plt.xlabel('Cultivo')
plt.ylabel('Área Sembrada (ha)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\n--- Fin del Análisis Exploratorio de Datos ---")
```
